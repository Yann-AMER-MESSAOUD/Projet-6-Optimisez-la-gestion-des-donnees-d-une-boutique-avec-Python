```python
#Création d'une colonne "stock_status_2
#La valeur de cette deuxième colonne sera fonction de la valeur dans la colonne "stock_quantity"
#si la valeur de la colonne "stock_quantity" est nulle renseigner "outofstock" sinon mettre "instock"
df_erp['stock_status_2'] = df_erp['stock_quantity'].apply(lambda x: 'outofstock' if x <= 0 else 'instock')
```

```python
#Vérifions que les 2 colonnes sont identiques:
#Les 2 colonnes sont strictement identiques si les valeurs de chaque ligne sont strictement identiques 2 à 2
#La comparaison de 2 colonnes peut se réaliser simplement avec l'instruction ci-dessous:
comparison = df_erp["stock_status"] == df_erp["stock_status_2"]
print("Nombre de lignes identiques: {}".format(comparison.sum()))
#Le résultat est l'affichage de True ou False pour chacune des lignes du dataset
print(comparison)
#C'est un bon début, mais difficile à exploiter
```

```python
#Mais il est possible de synthétiser ce résultat en effectuant la somme de cette colonne:
#True vaut 1 et False 0
#Nous devrions obtenir la somme de 824 qui correspond au nombre de lignes dans ce dataset
somme_comparison = comparison.sum()
print("Somme des lignes identiques : {}".format(somme_comparison))
```

```python
#Si les colonnes ne sont absolument pas identiques ligne à ligne alors identifier la ligne en écart
##Dans ce cas je vous ce lien pour apprendre à réaliser des filtres dans Pandas:
##https://bitbucket.org/hrojas/learn-pandas/src/master/
##Lesson 3
differences = df_erp[~comparison]
print("Lignes en écart :")
differences.head()
```

```python
#Corriger la ou les données incohérentes
df_erp.loc[(df_erp['stock_status'] == 'outofstock') & (df_erp['stock_quantity'] > 0), 'stock_status'] = 'instock'
df_erp.loc[(df_erp['stock_status'] == 'instock') & (df_erp['stock_quantity'] <= 0), 'stock_status'] = 'outofstock'
for index, row in differences.iterrows():
    if row['stock_quantity'] > 0:
        df_erp.at[index, 'stock_status'] = 'instock'
    else:
        df_erp.at[index, 'stock_status'] = 'outofstock'

#Verification en utilisant le même code que plus haut pour afficher les problemes
comparison = df_erp["stock_status"] == df_erp["stock_status_2"]
print("Nombre de lignes identiques après correction: {}".format(comparison.sum()))
# Recalculer la colonne stock_status_2
df_erp['stock_status_2'] = df_erp['stock_quantity'].apply(lambda x: 'outofstock' if x <= 0 else 'instock')

# Refaire la comparaison
comparison = df_erp["stock_status"] == df_erp["stock_status_2"]
print("Nombre de lignes identiques après correction: {}".format(comparison.sum()))

# Vérifier s'il reste des écarts
differences = df_erp[~comparison]
print("Lignes en écart après correction :")
print(differences)

```

```python
############################
# Analyse du taux de marge #
############################
# Création de la colonne prix HT
# On suppose que le taux de TVA est de 20%. Si le taux est différent, ajustez le calcul en conséquence.
df_final['prix_ht'] = df_final['price'] / 1.20

# Création de la colonne Taux de marge
# Taux de marge = (Prix HT - Prix d'achat) / Prix HT
df_final['taux_marge'] = (df_final['prix_ht'] - df_final['purchase_price']) / df_final['prix_ht']

# Afficher le prix minimum de la colonne "taux_marge"
min_taux_marge = df_final['taux_marge'].min()
print("Prix minimum du taux de marge :", min_taux_marge)

# Afficher le prix maximum de la colonne "taux_marge"
max_taux_marge = df_final['taux_marge'].max()
print("Prix maximum du taux de marge :", max_taux_marge)
```

```python
############################
# Analyse des correlations #
############################

#Importation de Seaborn
import seaborn as sns

# Supposons que df_final est déjà défini avec les colonnes nécessaires

# Calculer la matrice de corrélation pour les colonnes d'intérêt
corr_matrix = df_final[['stock_quantity', 'total_sales', 'price']].corr()

# Créer un masque pour n'afficher que la partie inférieure de la heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Création de la heatmap avec Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, 
            annot=True,                # Afficher les valeurs des corrélations
            cmap='coolwarm',           # Couleurs de la heatmap
            mask=mask,                 # Appliquer le masque pour n'afficher que la moitié
            vmin=-1, vmax=1,           # Plage des valeurs de la heatmap
            center=0,                  # Centre de la palette de couleurs
            square=True,               # Pour afficher la heatmap sous forme de carré
            linewidths=0.5,            # Largeur des lignes de séparation
            cbar_kws={"shrink": .8})   # Ajuster la taille de la barre de couleur
plt.title('Heatmap des Corrélations entre Stock, Ventes et Prix')
plt.tight_layout()
plt.show()
```
